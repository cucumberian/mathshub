#### Вопрос 12 (не по дата-сету)
```
В таблице нижепредставлены данные о товарообороте компаний в месяц. Необходимо определить: 

- средний размер месячного товарооборота на одну фирму
- моду и медиану месячного товарооборота
```

![123](pic1.png)

Таблица должна быть такая, иначе поиск всех значений мод и медиан вообще впринципе не представляю как произвести.

|№|Товарооборот в месяц, млн. р|Количество компаний|
|---|:---:|---:|
1 |  2.5 | 15 |
2 |  7.5 | 23 |
3 | 12.5 | 15 |
4 | 17.5 | 10 |
5 | 22.5 | 8 |


````python
import numpy as np
import statistics

a = [2.5 for i in range(15)]
a.extend([7.5 for i in range(23)])
a.extend([12.5 for i in range(15)])
a.extend([17.5 for i in range(10)])
a.extend([22.5 for i in range(8)])

print(f"mode = {statistics.mode(a)}")
print(f"median = {statistics.median(a)}")
print(f"mean = {statistics.mean(a):3.1f}")

# мода - самое часто встречающиеся значение - чаще всего встречается 23 раза значение от 5 до 10, среднее между ними 7.5

# медиана - значение посередине отсортированного датасета всего элементов 71, средний из них 36й, по порядку сначала идёт 15 по 2.5, затем 23 по 7.5 значений
# значит 36й элемент будет о значением 7.5

# среднее - расчитываем как среднеарифметическое от всех средних
mean = (15 * 2.5 + 23 * 7.5 + 15 * 12.5 + 10 * 17.5 + 8 * 22.5) / (15 + 23 + 15 + 10 + 8)
print(f"mean = {mean:3.1f}")
````
