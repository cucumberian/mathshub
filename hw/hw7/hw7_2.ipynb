```
В таблице нижепредставлены данные о товарообороте компаний в месяц. Необходимо определить: 

- средний размер месячного товарооборота на одну фирму
- моду и медиану месячного товарооборота
```

![123](pic1.png)

Есть формулы для определения значений интервального ряда.

https://math.semestr.ru/group/variations.php

````python
"""
Определяем медианный ряд.
Всего значений 71, среднее по счету 36, оно находится в интервале №2 - 5-10 с частотностью 23
"""
med = 5 + 5 * ( (15 + 23 + 15 + 10 + 8) / 2 - 15) / 23
med
````

````python
"""
Определяем моду
Ряд с модой будет также 23, так как в нём самая большая частотность.
"""

mode = 5 + 5 * (23-15) / ((23-15) + (23-15))
mode
````

````python
"""
Расчитывем среднее
Считал среднее в каждом ряду
|  2.5 | 15 |
|  7.5 | 23 |
| 12.5 | 15 |
| 17.5 | 10 |
| 22.5 |  8 |
Потом по формуле среднего M = xi * fi / sum(fi)
"""
mean = (2.5*15 + 7.5*23 + 12.5*15 + 17.5*10 + 22.5*8) / (15 + 23 + 15 + 10 + 8)
mean
````
