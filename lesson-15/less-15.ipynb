```python
import statistics   # библиотека для работы со статистикой
```

```python
tries_list = [5, 8, 9, 23, 12, 66, 3, 1, 1, 2, 3]   # набор значений
print(tries_list.__len__())
```

```python
# срежнее арифметическое для конечной выборки

mean = statistics.mean(tries_list)
print(f"{mean = :3.2f}")
```

```python
# медиана для дискретнго ряда
```

#### Медиана

- для нечетного количества элементов ищется серединное значений, половина значений ряда меньше медианы, половина значений больше.
- для четного количества элементов в выборке берутся два серединных значения и находится среднее между ними.

```python
median = statistics.median(tries_list)
print(f"{median = }")
```

#### Мода
Мода в дискретном ряду - значение, которое чаще всего встречается в статисчтическом ряду (вариант с наибольшей частотой).

Значния списка сортируются по возрастанию и для каждого элемента считается число его вхождений в множество.
- берется самое часто встречающиеся значение, самое частотное.
- если максимальная частотность двух элементов одинаково, и между элментами нет других (соседние или смежные элементы в упорядоченном ряду), то берется среднее значение между ними.
- если макимальная частотность для двух элментов максимальная, и они не стоят рядом, то мода будет __`бимодальной`__.
- аналогично с _бимодальной_ может быть и __`мультимодальная мода`__.

```python
print(sorted(tries_list))

import collections
counter = collections.Counter(tries_list)
print(counter)


mode = statistics.mode(tries_list)
print(f"{mode = }") # ответ 3, хотя мода у нас бимодальнаяb
# в версии питона до 3.7 велючительно поиск моды в мультимоддальном случае может не работать 


multimode = statistics.multimode(tries_list)    # работает начиная с версии python3.8
print(f"{multimode = }")
```
